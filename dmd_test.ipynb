```python
import numpy as np
from pydmd import BOPDMD

# 1. Prepare dummy spatio-temporal data
# X shape: (n_features/spatial_points, m_snapshots)
n_features = 50
m_snapshots = 100
X = np.sin(np.outer(np.linspace(0, 10, n_features), np.linspace(0, 5, m_snapshots)))
t = np.linspace(0, 5, m_snapshots)  # Historical time vector

# 2. Initialize BOPDMD
bopdmd = BOPDMD(
    svd_rank=10,        # Number of spatio-temporal modes to learn
    num_trials=100,     # Ensembling trials for bagging & uncertainty metrics
    trial_size=0.8      # Use 80% of data randomly per trial
)

# 3. Fit the model to historical data
bopdmd.fit(X, t)

# 4. Create a future time vector for predictions
# (e.g., forecasting 50 steps into the future with the same step size)
dt = t[1] - t[0]
t_future = np.linspace(t[-1] + dt, t[-1] + (50 * dt), 50)

# 5. Predict using the forecast method
forecast_mean, forecast_variance = bopdmd.forecast(t_future)

# Outputs:
# forecast_mean: shape (n_features, 50) -> The actual future predictions
# forecast_variance: shape (n_features, 50) -> The uncertainty at each point
```
